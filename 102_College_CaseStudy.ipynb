```python
#Bibliotheken importieren
import pandas as pd 
import numpy as np 
import matplotlib as mpl 
import seaborn as sns
```

```python
#mit der Pandas Funktion read_csv() wird aus der CSV Datei ein DataFrame erstellt
#Falls die csv Datei nicht im gleichen Verzeichnis wie das Python Script liegt, muss der Pfad angegeben werden
#Ansonsnten reicht der Dateiname
dataframe = pd.read_csv("102_College.csv")
#die pandas Funktion head() gibt die ersten 5 Zeilen des DataFrames aus
dataframe.head()
```

## Was ist ein DataFrame?
Ein DataFrame ist wie eine Tabelle mit Zeilen und Spalten in Python. Man kann es sich vorstellen wie eine Excel-Tabelle. Jede Spalte kann verschiedene Arten von Informationen speichern, wie Zahlen, Wörter oder Ja/Nein-Werte.

Mit einem DataFrame kann man viele Dinge tun. Man kann Daten sortieren, filtern, zusammenfassen und vieles mehr. Man kann auch Daten aus verschiedenen Quellen wie CSV-Dateien, Excel-Dateien oder Datenbanken in einen DataFrame laden und Daten aus einem DataFrame in diese Formate exportieren.

```python
#Die Pandas Funktion describe() gibt eine Zusammenfassung der statistischen Werte des DataFrames aus
dataframe.describe()
```

```python
#Die Pandas Funktion head() greift/wiedergibt die ersten Zeilen des DataFrames
#Die Anzahl der Zeilen kann als Parameter übergeben werden
#Der neuen Variable dataframe_10 wird der DataFrame mit den ersten 10 Zeilen zugewiesen
#somit ist dataframe_10 ein neuer DataFrame
dataframe_10 = dataframe.head(10)
dataframe_10
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

scatter_matrix(dataframe_10, figsize=(10,10))
```

## Wozu dient Scatter Matrix

Ein Scatter-Matrix (Streu-Matrix) ist wie ein Diagramm, das uns hilft, die Beziehung zwischen verschiedenen Datenpunkten zu sehen. Es zeigt uns, ob und wie verschiedene Daten zusammenhängen. Es ist wie ein Diagramm, das viele kleine Diagramme enthält, jedes zeigt, wie zwei Datenpunkte zusammenhängen. Es ist sehr nützlich, um Muster in den Daten zu erkennen.


```python
#figsize gibt die Größe des Plots an
#s gibt die Größe der Punkte an
#marker gibt die Form der Punkte an
scatter_matrix(dataframe_10, figsize=(15,15), s=50, marker='D')
```

```python
from pandas.plotting import boxplot
boxplot(dataframe, "Outstate", figsize=(15,15))
```

## Interpretiere den Boxplot

Die "unteren" 25% der Daten-Einträge liegen dichter an einander als die "oberen" 25%, mit Werten zwischen 2.400€ und 7.400€
