## Was ist ein DataFrame?
Ein DataFrame ist wie eine Tabelle mit Zeilen und Spalten in Python. Man kann es sich vorstellen wie eine Excel-Tabelle. Jede Spalte kann verschiedene Arten von Informationen speichern, wie Zahlen, WÃ¶rter oder Ja/Nein-Werte.

Mit einem DataFrame kann man viele Dinge tun. Man kann Daten sortieren, filtern, zusammenfassen und vieles mehr. Man kann auch Daten aus verschiedenen Quellen wie CSV-Dateien, Excel-Dateien oder Datenbanken in einen DataFrame laden und Daten aus einem DataFrame in diese Formate exportieren.

## Wozu dient Scatter Matrix

Ein Scatter-Matrix (Streu-Matrix) ist wie ein Diagramm, das uns hilft, die Beziehung zwischen verschiedenen Datenpunkten zu sehen. Es zeigt uns, ob und wie verschiedene Daten zusammenhÃ¤ngen. Es ist wie ein Diagramm, das viele kleine Diagramme enthÃ¤lt, jedes zeigt, wie zwei Datenpunkte zusammenhÃ¤ngen. Es ist sehr nÃ¼tzlich, um Muster in den Daten zu erkennen.


```python
#figsize gibt die GrÃ¶Ãe des Plots an
#s gibt die GrÃ¶Ãe der Punkte an
#marker gibt die Form der Punkte an
scatter_matrix(dataframe_10, figsize=(15,15), s=50, marker='D')
```
